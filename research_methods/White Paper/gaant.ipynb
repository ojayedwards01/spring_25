```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
import seaborn as sns

# Define the timeline for a 1-year project starting May 2025
tasks = [
    ("Literature Review & Dataset Survey", "2025-05-01", "2025-06-15"),
    ("Data Cleaning & Preprocessing", "2025-06-16", "2025-07-15"),
    ("Fine-tuning ASR Models (Whisper)", "2025-07-16", "2025-08-31"),
    ("Fine-tuning MT Models (mBART/NLLB)", "2025-09-01", "2025-10-15"),
    ("Training TTS Models (Coqui)", "2025-10-16", "2025-11-30"),
    ("Pipeline Integration", "2025-12-01", "2026-01-15"),
    ("Model Evaluation & User Studies", "2026-01-16", "2026-02-28"),  # Changed end date to 2026-02-28
    ("Final Report & Deployment", "2026-03-01", "2026-04-15")
]

# Convert to DataFrame
df = pd.DataFrame(tasks, columns=["Task", "Start", "End"])
df["Start"] = pd.to_datetime(df["Start"])
df["End"] = pd.to_datetime(df["End"])

# Plot Gantt Chart
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
colors = sns.color_palette("husl", len(df))

for i, task in df.iterrows():
    plt.barh(task["Task"], (task["End"] - task["Start"]).days, left=task["Start"], color=colors[i])

plt.xlabel("Timeline")
plt.ylabel("Project Tasks")
# plt.title("Research Timeline: Multilingual Speech Translation in Nigeria (May 2025  Apr 2026)")
plt.title("Leveraging Community Speech Corpora to Build a Multilingual Translation Pipeline in Nigeria (May 2025  Apr 2026)")
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
plt.tight_layout()
plt.grid(True, axis='x')
plt.xticks(rotation=45)

plt.savefig("Gantt_Chart_Multilingual_Speech_Project.png")
plt.show()
```
